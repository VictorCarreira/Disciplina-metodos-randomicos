```python
################PACOTES################
import numpy as np
import pylab as py
import matplotlib
import matplotlib.pyplot as plt
#####################################
```

```python
fit=np.loadtxt('fit.txt')
#print(fit)
rms=np.loadtxt('rrms.txt')
#print(rms)
```

```python
#Gerando o Grfico do ajuste:

x=fit[:,0]
y=fit[:,1]
z=fit[:,2]
e=rms
error = []
for i in range(len(x)):
    error.append(e)

py.rcParams['figure.figsize'] = (20.0, 10.0)#Redimensiona a figura

plt.plot( x, y, 'ko-',label="D.obs") 
plt.plot( x, z, 'ro-',label="D.calc") 
plt.plot(x,error,'bo-', label=e)
plt.plot(e,'-')

legend = plt.legend(loc='upper right', shadow=True, fontsize='x-large')

plt.title("Esfera Magnetizada Enterrada", fontweight='bold', fontsize=25)
plt.tick_params(axis='both', labelsize=15)
#plt.errorbar(x, z, yerr=e,label="RMS")

plt.grid(True)
plt.xlabel("$metros$", fontsize=15)
plt.ylabel("$nT$", fontsize=15)
plt.savefig('Ajuste.png')  
plt.show()  
plt.close()
```
